```python
#wapp to sort the elements in acending order using bubble, selection and insertion sort
def bubblesort(lis):
    for n in range(len(lis) - 1, 0, -1):

        # Inner loop to compare adjacent elements
        for i in range(n):
            if lis[i] > lis[i + 1]:

                # Swap elements if they are in the wrong order
                swapped = True
                lis[i], lis[i + 1] = lis[i + 1], lis[i]
    return lis

def selectionsort(lis):
    for i in range(len(lis)):
        mini = lis[i]
        flag = 0
        for j in range(i,len(lis)):
            if lis[j]<mini:
                mini = lis[j]
                minpos = j
                flag = 1
        if flag:
            temp = lis[minpos]
            lis[minpos] = lis[i]
            lis[i] = temp
    return lis

def insertionsort(lis):
    n = len(lis)  # Get the length of the lisay
    if n <= 1:
        return lis
    for i in range(1, n): 
        key = lis[i]  
        j = i-1
        while j >= 0 and key < lis[j]:  
            lis[j+1] = lis[j]  
            j -= 1
        lis[j+1] = key  
    return lis
n = int(input("How many numbers? "))
lis=[]
for i in range(n):
    x = int(input("Enter a number "))
    lis.append(x)
print(f"Bubble sort {bubblesort(lis)}")
print(f"Selection sort {selectionsort(lis)}")
print(f"Insertion sort {insertionsort(lis)}")
```

```python
#wapp to sort the elements in acending order using merge sort
def merge(lis1,lis2):
    i = j = 0
    res = []
    while(i<len(lis1) and j<len(lis2)):
        if lis1[i]<lis2[j]:
            res.append(lis1[i])
            i+=1
        elif lis1[i]>= lis2[j]:
            res.append(lis2[j])
            j+=1
    
    while(j<len(lis2)):
        res.append(lis2[j])
        j+= 1
    while(i<len(lis1)):
        res.append(lis1[i])
        i += 1
    return res

def merge_sort(lis):
    if len(lis) <= 1:  # base case
        return lis

    half = len(lis) // 2
    left = merge_sort(lis[:half])
    right = merge_sort(lis[half:])

    return merge(left, right)
n = int(input("How many numbers? "))
lis=[]
for i in range(n):
    x = int(input("Enter a number "))
    lis.append(x)
print("Unsorted list ",lis)
print("Sorted list ",merge_sort(lis))
```

```python
#wap function that finds a key element through binary search and return it's position

def bubblesort(lis):
    for i in range(len(lis)):
        for j in range(i+1,len(lis)):
            if lis[j]<lis[i]:
                temp = lis[i]
                lis[i] = lis[j]
                lis[j] = temp
    return lis

def binary_search(lis, low, high, x):


    if high >= low:

        mid = (high + low) // 2

        
        if lis[mid] == x:
            return mid

        elif lis[mid] > x:
            return binary_search(lis, low, mid - 1, x)

        else:
            return binary_search(lis, mid + 1, high, x)

    else:
        return -1

n = int(input("How many numbers? "))
lis=[]
for i in range(n):
    x = int(input("Enter a number "))
    lis.append(x)
lis = bubblesort(lis)
print(lis)
x = int(input("Which element you want to find "))

# Function call
result = binary_search(lis, 0, len(lis)-1, x)

if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in list")
```

```python
#use linear search ^
n = int(input("How many numbers? "))
lis=[]
for i in range(n):
    x = int(input("Enter a number "))
    lis.append(x)
def linear_search(lis,key):
    flag = 0
    for i in range(len(lis)):
        if lis[i] == key:
            flag = 1
            pos = i
            break
    if flag:
        return pos
    return -1
print(lis)

k = int(input("Enter the element u want to search "))
p = linear_search(lis,k)
if p!= -1:
    print(f"The position of {k} is {p}")
else:
    print(f"{k} is not in the list")
```

```python
#wapp to implement stack ops, give stack size first
msize = int(input("Enter stack size"))
stack = []
def spop(stack):
    if len(stack) == 0:
        print("stack underflow")
        return
    stack.pop()
def spush(stack,data):
    if len(stack) == msize:
        print("Stack overflow")
    else:
        stack.append(data)
while(True):
    choice = int(input("enter 1 to push 2 to pop 3 to print stack anything else will exit :"))
    if choice == 1:
        data = int(input("Enter a number "))
        spush(stack,data)
    elif choice == 2:
        spop(stack)
        print("done")
    elif choice == 3:
        print(stack)
    else:
        print("QUITTING")
        break
```

```python
#wapp to implement queue ops
msize = int(input("Enter queue size"))
queue = []
def spop(queue):
    if len(queue) == 0:
        print("queue underflow")
        return
    queue.pop(0)
def spush(queue,data):
    if len(queue) == msize:
        print("queue overflow")
    else:
        queue.append(data)
while(True):
    choice = int(input("enter 1 to push 2 to pop 3 to print queue anything else will exit :"))
    if choice == 1:
        data = int(input("Enter a number "))
        spush(queue,data)
    elif choice == 2:
        spop(queue)
        print("done")
    elif choice == 3:
        print(queue)
    else:
        print("QUITTING")
        break
```
